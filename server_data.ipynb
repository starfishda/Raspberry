```python
import requests
import csv
import pandas as pd

def upload():
    try:
        line = ""
        f = open("data.csv", 'r')
        for tmp in f:
            line += tmp
            print(line)
        f.close()
        data = {'data': line}
        myUrl = 'http://wiffy.io/sen/upload.php'
        requests.post(myUrl, data=data).text
    except:
        print('ERROR')

def download():
    try:
        myUrl = 'http://wiffy.io/sen/data.csv'
        res = requests.post(myUrl).text
        
        sres = res.split("\n")
        f = open("data.csv", 'w')
        for tmp in sres:
            f.write(tmp + "\n")
        f.close()
    except:
        print('ERROR')

def delete():
    try:
        myUrl = 'http://wiffy.io/sen/delete.php'
        requests.post(myUrl).text
    except:
        print('ERROR')
        
list = ['2019-12-02 09:20:53.861315', 143, 126, 840, 453, 270]
download()
df = pd.read_csv("data.csv")
#df.loc[len(df)] = [n for n in list]
#df.to_csv("data.csv", index = False)
#df.to_csv("data.csv", mode='w', index = False)
#upload()
df
```
